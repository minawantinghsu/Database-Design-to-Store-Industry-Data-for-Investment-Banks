```python
print('Connecting to the PostgreSQL database...')
conn = psycopg2.connect(
    host="localhost",
    port='5432',
    database="5400",
    user="postgres",
    password="123")
```

```python
cur = conn.cursor()
```

```python
cur.execute(
        """
 CREATE TABLE Capacity(
 location varchar PRIMARY KEY,
 capacity numeric,
 effective_capacity numeric);
 
 CREATE TABLE Cost_Margin (
 cost_margin_id int,
 location varchar,
 date varchar,
 gross_margin numeric,
 cost numeric,
 PRIMARY KEY(cost_margin_id),
 FOREIGN KEY(location) REFERENCES Capacity);

 CREATE TABLE Export(
 export_id int,
 location varchar,
 date varchar,
 export_amount numeric,
 PRIMARY KEY(export_id),
 FOREIGN KEY(location) REFERENCES Capacity);

 CREATE TABLE Sold_price(
 sold_price_id int,
 location varchar,
 date varchar,
 level varchar,
 price int,
 PRIMARY KEY(sold_price_id),
 FOREIGN KEY(location) REFERENCES Capacity);
 
""")
    
conn.commit()
```
